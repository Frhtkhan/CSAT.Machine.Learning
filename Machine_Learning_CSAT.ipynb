````python
# Create copy of dataset
df1 = df.copy()

# Check duplicate values
print("Duplicate Values :", df1.duplicated().sum())

# Remove duplicate values
df1 = df1.drop_duplicates()

# Check missing values
print("\nMissing Values:\n")
print(df1.isnull().sum())

# Handling missing values
# Numerical columns
num_cols = df1.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df1[col].fillna(df1[col].median(), inplace=True)

# Categorical columns
cat_cols = df1.select_dtypes(include=['object']).columns

for col in cat_cols:
    df1[col].fillna(df1[col].mode()[0], inplace=True)

# Check dataset shape after cleaning
print("\nDataset Shape After Cleaning :", df1.shape)

# Final missing values check
print("\nRemaining Missing Values:\n")
print(df1.isnull().sum())
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothetical Statements
Customer support channels significantly affect CSAT scores.
Higher connected handling time leads to lower customer satisfaction scores.
Different product categories have significantly different average CSAT scores.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Customer support channels significantly affect CSAT scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import f_oneway

# Group CSAT scores by channel
groups = [
    group['CSAT Score'].values
    for name, group in df.groupby('channel_name')
]

# Perform ANOVA Test
f_stat, p_value = f_oneway(*groups)

print("P-Value :", p_value)
````

##### Which statistical test have you done to obtain P-Value?

We used the One-Way ANOVA Test to obtain the p-value because we are comparing the mean CSAT scores across multiple customer support channels.

##### Why did you choose the specific statistical test?

We chose the One-Way ANOVA Test because it is used to compare the mean values of one numerical variable (CSAT Score) across more than two categorical groups (customer support channels).

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Higher connected handling time leads to lower customer satisfaction (CSAT) scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Remove missing values
temp_df = df[['connected_handling_time', 'CSAT Score']].dropna()

# Perform Pearson Correlation Test
corr, p_value = pearsonr(
    temp_df['connected_handling_time'],
    temp_df['CSAT Score']
)

print("Correlation :", corr)
print("P-Value :", p_value)
````

##### Which statistical test have you done to obtain P-Value?



We used the Pearson Correlation Test to obtain the p-value because we wanted to measure the relationship between two numerical variables: connected handling time and CSAT score.

##### Why did you choose the specific statistical test?

We chose the Pearson Correlation Test because it is used to measure the strength and direction of the relationship between two continuous numerical variables. Here, both connected handling time and CSAT score are numerical variables.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Different product categories have significantly different average CSAT scores.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Group CSAT scores by Product Category
groups = [
    group['CSAT Score'].values
    for name, group in df.groupby('Product_category')
]

# Perform ANOVA Test
f_stat, p_value = f_oneway(*groups)

print("P-Value :", p_value)
````

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
# Check missing values
df.isnull().sum()

# Filling missing values in numerical columns using median
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Filling missing values in categorical columns using mode
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Verify missing values after imputation
df.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

We used median imputation for numerical columns and mode imputation for categorical columns. Median was used because it handles outliers effectively, while mode was used because it replaces missing categorical values with the most frequently occurring category.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
num_cols = df.select_dtypes(include=['int64','float64']).columns

# Detecting and treating outliers using IQR method
for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping outliers
    df[col] = df[col].clip(lower_bound, upper_bound)

print("Outlier treatment completed.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

We used the IQR method to handle outliers by capping extreme values. This technique was chosen because it reduces the effect of abnormal data points and improves model accuracy without removing important data.

#### 6. Rephrase Text

````python
# Rephrase Text
import re
import string
import contractions
import nltk

from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')

# Example text column
text_col = 'Customer Remarks'

# Load stopwords
stop_words = set(stopwords.words('english'))

# Text Cleaning Function
def clean_text(text):

    text = str(text).lower()

    # Expand contractions
    text = contractions.fix(text)

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove words containing digits
    text = re.sub(r'\w*\d\w*', '', text)

    # Remove stopwords
    text = ' '.join(
        word for word in text.split()
        if word not in stop_words
    )

    # Remove extra white spaces
    text = " ".join(text.split())

    return text

# Apply cleaning function
df[text_col] = df[text_col].apply(clean_text)

# View cleaned text
df[text_col].head()
````
